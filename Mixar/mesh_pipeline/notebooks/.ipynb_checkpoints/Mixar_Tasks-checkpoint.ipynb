## TASK 2 â€” Normalization & Quantization

**Deliverables shown here**
- Two normalized meshes (MinMax + UnitSphere) saved to `outputs/normalized/`.
- Two quantized-visualizable meshes saved to `outputs/quantized_vis/`.
- Short printed comparison of which normalization preserves structure better (based on MSE after quantization round-trip).

We will:
1. Normalize original vertices with both methods.
2. Save normalized meshes as `.obj`.
3. Quantize (n_bins=1024), dequantize, denormalize, save reconstructions for visual check.
4. Compute and print reconstruction MSE/MAE for both normalizers.


```python
# TASK 2: export normalized, quantized_vis, compute recon errors for both normalizers
mesh_info = load_mesh_info(mesh_path)
verts = mesh_info["vertices"]
faces = mesh_info["faces"]

# folders
OUT_N = OUT_ROOT / "normalized"
OUT_QV = OUT_ROOT / "quantized_vis"
OUT_REC = OUT_ROOT / "reconstructed"
OUT_N.mkdir(parents=True, exist_ok=True)
OUT_QV.mkdir(parents=True, exist_ok=True)
OUT_REC.mkdir(parents=True, exist_ok=True)

def save_obj_vertices(vs, faces, out_path):
    out_path.parent.mkdir(parents=True, exist_ok=True)
    with open(out_path, "w", encoding="utf-8") as fh:
        for v in vs:
            fh.write(f"v {float(v[0])} {float(v[1])} {float(v[2])}\n")
        if faces is not None:
            for f in faces:
                fh.write(f"f {int(f[0])+1} {int(f[1])+1} {int(f[2])+1}\n")

# 1) MinMax
mn = MinMaxNormalizer().fit(verts)
v_minmax = mn.transform(verts)   # [0,1]
save_obj_vertices(v_minmax, faces, OUT_N / "minmax" / f"{mesh_path.stem}_norm.obj")

# quantize -> dequantize -> denorm -> save recon (visualizable)
n_bins = 1024
q = Quantizer(n_bins=n_bins).transform(v_minmax)
v_deq = Quantizer(n_bins=n_bins).inverse_transform(q)
v_recon_minmax = mn.inverse_transform(v_deq)
save_obj_vertices(v_recon_minmax, faces, OUT_QV / "minmax" / f"{mesh_path.stem}_q_vis.obj")
save_mesh(trimesh.Trimesh(vertices=v_recon_minmax, faces=faces), str(OUT_REC / f"{mesh_path.stem}_minmax_recon.obj"))

err_min = compute_errors(verts, v_recon_minmax)
print("MinMax recon error:", err_min)

# 2) UnitSphere
us = UnitSphereNormalizer().fit(verts)
v_us = us.transform(verts)   # roughly [-1,1]
save_obj_vertices(v_us, faces, OUT_N / "unitsphere" / f"{mesh_path.stem}_norm.obj")

# map to [0,1] for quantization, then back
v_us_mapped = (v_us + 1.0) / 2.0
q2 = Quantizer(n_bins=n_bins).transform(v_us_mapped)
v_deq2 = Quantizer(n_bins=n_bins).inverse_transform(q2)
v_us_recon_mapped = v_deq2
v_us_recon = us.inverse_transform(v_us_recon_mapped)
save_obj_vertices(v_us_recon, faces, OUT_QV / "unitsphere" / f"{mesh_path.stem}_q_vis.obj")
save_mesh(trimesh.Trimesh(vertices=v_us_recon, faces=faces), str(OUT_REC / f"{mesh_path.stem}_unitsphere_recon.obj"))

err_us = compute_errors(verts, v_us_recon)
print("UnitSphere recon error:", err_us)

# Short written comparison printed
print("\nShort comparison:")
if err_min["mse"] < err_us["mse"]:
    print("MinMax normalization gives lower MSE on this mesh (better preservation).")
else:
    print("UnitSphere normalization gives lower MSE on this mesh (better preservation).")

# show the quantized-visualizable recon (minmax) inline if available
img_compare = OUT_ROOT / "metrics" / f"{mesh_path.stem}_error.png"
# if compare image exists show it; else show HQ recon
try:
    from src.viz import compare_meshes_plot
    # create side by side compare image of original and minmax reconstruction
    orig_mesh = trimesh.Trimesh(vertices=verts, faces=faces)
    rec_mesh = trimesh.Trimesh(vertices=v_recon_minmax, faces=faces)
    cmp_out = OUT_ROOT / "renders" / f"{mesh_path.stem}_compare_minmax.png"
    cmp_out.parent.mkdir(parents=True, exist_ok=True)
    compare_meshes_plot(orig_mesh, rec_mesh, str(cmp_out))
    display(Image(str(cmp_out)))
except Exception as e:
    print("Compare render failed:", e)
```

## Adaptive Bonus â€” Rotation invariance & adaptive quantization

**Deliverables**
- Normalized & quantized outputs for multiple orientations (saved under `outputs/adaptive_bonus/`).
- Error plots comparing uniform vs adaptive quantization over rotations.
- A comparison table summarizing reconstruction error across methods.
- Short written analysis discussing invariance and adaptive effectiveness.

We run adaptive experiment for N rotations (default 8) and produce the per-mesh summary JSONs.


```python
def vertex_density(vertices, k=16):
    vertices = np.asarray(vertices, float)
    if vertices.shape[0] <= k:
        mean_d = np.linalg.norm(vertices - vertices.mean(axis=0), axis=1)
        return 1.0 / (mean_d + 1e-12), mean_d
    tree = cKDTree(vertices)
    dists, _ = tree.query(vertices, k=k+1)
    mean_d = dists[:, 1:].mean(axis=1)
    return 1.0 / (mean_d + 1e-12), mean_d

def adaptive_quantize_roundtrip(vertices, faces, bins_low=256, bins_med=1024, bins_high=4096, k=16):
    mn = MinMaxNormalizer().fit(vertices)
    v_norm = mn.transform(vertices)
    dens, _ = vertex_density(vertices, k=k)
    dmin, dmax = float(dens.min()), float(dens.max())
    dens_norm = (dens - dmin) / (dmax - dmin + 1e-12)
    q1, q2 = np.quantile(dens_norm, [0.33, 0.66])
    bins_per_vertex = np.full(len(vertices), bins_med, int)
    bins_per_vertex[dens_norm <= q1] = bins_low
    bins_per_vertex[dens_norm > q2] = bins_high
    q_all = np.zeros_like(v_norm, int)
    deq_all = np.zeros_like(v_norm, float)
    for b in np.unique(bins_per_vertex):
        idx = np.where(bins_per_vertex == b)[0]
        if idx.size == 0: continue
        v_sub = v_norm[idx]
        q_sub = np.floor(v_sub * (b - 1)).astype(int)
        q_sub = np.clip(q_sub, 0, b - 1)
        deq_sub = q_sub.astype(float) / (b - 1)
        q_all[idx] = q_sub
        deq_all[idx] = deq_sub
    recon_vertices = mn.inverse_transform(deq_all)
    return recon_vertices

def eval_adaptive_vs_uniform(mesh_path, n_rot=8, random_seed=42):
    np.random.seed(random_seed)
    mesh = trimesh.load(str(mesh_path), process=False)
    faces = getattr(mesh, "faces", None)
    results = {"uniform": [], "adaptive": []}
    for i in range(n_rot):
        R = trimesh.transformations.random_rotation_matrix()
        m2 = mesh.copy()
        m2.apply_transform(R)
        verts = m2.vertices
        # uniform
        mn = MinMaxNormalizer().fit(verts)
        v_norm = mn.transform(verts)
        q = Quantizer(n_bins=1024).transform(v_norm)
        v_deq = Quantizer(n_bins=1024).inverse_transform(q)
        recon_uniform = mn.inverse_transform(v_deq)
        err_u = compute_errors(verts, recon_uniform)
        results["uniform"].append(float(err_u["mse"]))
        # adaptive
        recon_adaptive = adaptive_quantize_roundtrip(verts, faces, 256, 1024, 4096, 16)
        err_a = compute_errors(verts, recon_adaptive)
        results["adaptive"].append(float(err_a["mse"]))
    return {
        "mesh": mesh_path.name,
        "uniform_mse_mean": float(np.mean(results["uniform"])),
        "adaptive_mse_mean": float(np.mean(results["adaptive"])),
        "uniform_mse_list": results["uniform"],
        "adaptive_mse_list": results["adaptive"]
    }
```

```python
# Adaptive bonus experiment: run for one mesh (branch.obj) or all meshes
# If eval_adaptive_vs_uniform is not inside src.pipeline, we fall back to the notebook function defined earlier.
try:
    # prefer pipeline wrapper if implemented
    summary = process_mesh_adaptive(mesh_path, OUT_ROOT, bins_low=256, bins_med=1024, bins_high=4096, k=16, n_rot=8)
    print("Adaptive summary saved by process_mesh_adaptive (pipeline).")
except Exception as e:
    print("process_mesh_adaptive not available or failed:", e)
    # fallback: use eval_adaptive_vs_uniform from notebook (we assumed it's defined earlier)
    try:
        summary = eval_adaptive_vs_uniform(mesh_path, n_rot=8)
        (OUT_ROOT / "adaptive_bonus").mkdir(parents=True, exist_ok=True)
        (OUT_ROOT / "adaptive_bonus" / f"{mesh_path.stem}_adaptive_summary.json").write_text(json.dumps(summary, indent=2))
        print("Adaptive summary (notebook eval) saved.")
    except Exception as e2:
        print("Fallback adaptive eval failed:", e2)
        summary = None

if summary:
    print("Adaptive summary (printed):")
    print(json.dumps(summary, indent=2))

# Create a small comparison table to display in notebook
if summary:
    from IPython.display import HTML, display
    rows = [
        ("Method", "Mean MSE"),
        ("Uniform (1024)", f"{summary['uniform_mse_mean']:.3e}"),
        ("Adaptive", f"{summary['adaptive_mse_mean']:.3e}")
    ]
    html = "<table><tr><th>Method</th><th>Mean MSE</th></tr>"
    html += f"<tr><td>Uniform (1024)</td><td>{summary['uniform_mse_mean']:.3e}</td></tr>"
    html += f"<tr><td>Adaptive (prototype)</td><td>{summary['adaptive_mse_mean']:.3e}</td></tr>"
    html += "</table>"
    display(HTML(html))

    # Short written analysis for notebook
    print("\nAdaptive analysis (notebook):")
    print(" - Uniform mean-MSE:", summary['uniform_mse_mean'])
    print(" - Adaptive mean-MSE:", summary['adaptive_mse_mean'])
    if summary['adaptive_mse_mean'] < summary['uniform_mse_mean']:
        print("Adaptive prototype improved MSE for this mesh (good).")
    else:
        print("Adaptive prototype did not improve MSE for this mesh (uniform is better).")
```
